```python
import numpy as np
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import StratifiedKFold, cross_val_score

oX = original_data["x"]
oy = original_data["y"]

store_max = {}

# y has to be 1D because right now it includes index
if len(oy.shape) == 2:
    oy = oy.iloc[:, 0]
oy = oy.astype(int)

real_y0 = oX.loc[oy == 0]
real_y1 = oX.loc[oy == 1]
syn_y0 = cvae_y0
syn_y1 = cvae_y1

syn_X = pd.concat([syn_y0, syn_y1], axis=0, ignore_index=True)

for col in oX.columns:
    q99 = oX[col].quantile(0.99)
    max_val = oX[col].max()
    store_max[col] = max_val
    # max approximately equals 99th percentile
    if abs(max_val - q99) < 1e-6: 
        print(col, max_val)

```

```python
from matplotlib import pyplot as plt


def summarize_feature(real: pd.Series, syn: pd.Series, ceiling=None, name=""):
    qs = [0.5, 0.9, 0.95, 0.99]
    out = {
        "feature": name,
        "real_mean": real.mean(),
        "syn_mean": syn.mean(),
        "real_std": real.std(),
        "syn_std": syn.std(),
    }
    for q in qs:
        out[f"real_q{int(q*100)}"] = real.quantile(q)
        out[f"syn_q{int(q*100)}"] = syn.quantile(q)
    if ceiling is not None:
        out["real_p_at_ceiling"] = (real == ceiling).mean()
        out["syn_p_at_ceiling"] = (syn == ceiling).mean()
    return out

import numpy as np

def plot_feature(real, syn, title="", bins=60, ceiling=None):
    plt.figure()

    all_vals = np.r_[real.values, syn.values]
    edges = np.histogram_bin_edges(all_vals, bins=bins)

    plt.hist(real.values, bins=edges, alpha=0.5, density=True, label="real")
    plt.hist(syn.values,  bins=edges, alpha=0.5, density=True, label="synthetic")
    if ceiling is not None:
        plt.axvline(ceiling, linestyle="--")

    plt.title(title)
    plt.xlabel("Value")
    plt.ylabel("Denisties")
    plt.legend()
    plt.show()

feat = "V10_blood_IgGRBD"
CEILING = store_max[feat] 

# y=0
summ_y0 = summarize_feature(real_y0[feat], syn_y0[feat], ceiling=CEILING, name=f"{feat} (y=0)")
# y=1
summ_y1 = summarize_feature(real_y1[feat], syn_y1[feat], ceiling=CEILING, name=f"{feat} (y=1)")

display(pd.DataFrame([summ_y0, summ_y1]))

plot_feature(
    real_y0[feat],
    syn_y0[feat],
    title=f"{feat} Distribution (y=0)",
    ceiling=CEILING
)

plot_feature(
    real_y1[feat],
    syn_y1[feat],
    title=f"{feat} Distribution (y=1)",
    ceiling=CEILING
)
```

```python

def ensure_df(X, columns):
    """Make sure X is a DataFrame with the given columns."""
    if isinstance(X, pd.DataFrame):
        return X[columns].copy()
    return pd.DataFrame(X, columns=columns)

def rf_auc_probe_binary(real_df: pd.DataFrame, syn_df: pd.DataFrame, seed=42, n_splits=5):
    """
    RF probe: label s=0 real, s=1 synthetic.
    Returns mean/std AUC via stratified CV.
    """
    n = min(len(real_df), len(syn_df))
    rng = np.random.default_rng(seed)

    # sample same n from each
    real_idx = rng.choice(real_df.index.to_numpy(), size=n, replace=False)
    syn_idx  = rng.choice(syn_df.index.to_numpy(),  size=n, replace=False)

    real_s = real_df.loc[real_idx]
    syn_s  = syn_df.loc[syn_idx]

    X = pd.concat([real_s, syn_s], axis=0).to_numpy()
    s = np.r_[np.zeros(n, dtype=int), np.ones(n, dtype=int)]

    rf = RandomForestClassifier(
        n_estimators=500,
        max_depth=None,
        random_state=seed,
        n_jobs=-1,
    )

    cv = StratifiedKFold(n_splits=n_splits, shuffle=True, random_state=seed)
    aucs = cross_val_score(rf, X, s, cv=cv, scoring="roc_auc")

    return float(aucs.mean()), float(aucs.std()), aucs

# Per-stratum probes 

cols = oX.columns

syn_y0_df = ensure_df(syn_y0, cols)
syn_y1_df = ensure_df(syn_y1, cols)

real_y0_df = ensure_df(real_y0, cols)
real_y1_df = ensure_df(real_y1, cols)

auc0_mean, auc0_std, auc0_folds = rf_auc_probe_binary(real_y0_df, syn_y0_df, seed=42)
auc1_mean, auc1_std, auc1_folds = rf_auc_probe_binary(real_y1_df, syn_y1_df, seed=42)

print(f"RF AUC (y=0): {auc0_mean:.3f} ± {auc0_std:.3f}   folds={np.round(auc0_folds,3)}")
print(f"RF AUC (y=1): {auc1_mean:.3f} ± {auc1_std:.3f}   folds={np.round(auc1_folds,3)}")
```

# Ablation section

```python
real_X = oX
real_y = oy

syn_X = pd.concat([syn_y0, syn_y1], axis=0, ignore_index=True)
syn_y = np.r_[
    np.zeros(len(syn_y0), dtype=int),
    np.ones(len(syn_y1), dtype=int)
]

# Balanced counts for probe
n_neg = min((real_y == 0).sum(), (syn_y == 0).sum())
n_pos = min((real_y == 1).sum(), (syn_y == 1).sum())
```

```python


rng = np.random.default_rng(42)

def strat_samp(neg_i, posi_i, n_neg, n_posi):
    chosen_neg = rng.choice(neg_i, size=n_neg, replace=False)
    chosen_posi = rng.choice(posi_i, size=n_posi, replace=False)
    return np.concatenate([chosen_neg, chosen_posi])

def rf_auc_probe_stratified(real_X: pd.DataFrame, real_y,
                             syn_X: pd.DataFrame, syn_y,
                             n_neg, n_pos, seed=42):

    real_y = np.asarray(real_y).astype(int)
    syn_y  = np.asarray(syn_y).astype(int)

    # indices per stratum
    real_neg_i = np.where(real_y == 0)[0]
    real_pos_i = np.where(real_y == 1)[0]

    syn_neg_i = np.where(syn_y == 0)[0]
    syn_pos_i = np.where(syn_y == 1)[0]

    # stratified sampling
    real_idx = strat_samp(real_neg_i, real_pos_i, n_neg, n_pos)
    syn_idx  = strat_samp(syn_neg_i,  syn_pos_i,  n_neg, n_pos)

    real_s = real_X.iloc[real_idx]
    syn_s  = syn_X.iloc[syn_idx]

    X = pd.concat([real_s, syn_s], axis=0).to_numpy()
    s = np.r_[np.zeros(len(real_s), dtype=int),
              np.ones(len(syn_s), dtype=int)]

    rf = RandomForestClassifier(
        n_estimators=500,
        random_state=seed,
        n_jobs=-1
    )

    cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=seed)
    aucs = cross_val_score(rf, X, s, cv=cv, scoring="roc_auc")

    return float(aucs.mean()), float(aucs.std()), aucs
```

# Ablation function

```python
def run_ablation(real_X, real_y, syn_X, syn_y, drop_feats, n_neg, n_pos, seed=42, label=""):
    # keep only features that exist
    drop_feats = [f for f in drop_feats if f in real_X.columns]

    real_ab = real_X.drop(columns=drop_feats)
    syn_ab  = syn_X.drop(columns=drop_feats)

    auc_mean, auc_std, folds = rf_auc_probe_stratified(
        real_X=real_ab,
        real_y=real_y,
        syn_X=syn_ab,
        syn_y=syn_y,
        n_neg=n_neg,
        n_pos=n_pos,
        seed=seed
    )

    print(f"[{label}] dropped {len(drop_feats)} features")
    print(f"  AUC: {auc_mean:.3f} ± {auc_std:.3f}  folds={np.round(folds,3)}")

    return auc_mean, auc_std, folds, drop_feats
```

# RF-driven ablation

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import roc_auc_score

def fit_rf_get_importance(real_X, real_y, syn_X, syn_y, n_neg, n_pos, seed=42):
    real_y = np.asarray(real_y).astype(int)
    syn_y  = np.asarray(syn_y).astype(int)

    real_neg_i = np.where(real_y == 0)[0]
    real_pos_i = np.where(real_y == 1)[0]
    syn_neg_i  = np.where(syn_y == 0)[0]
    syn_pos_i  = np.where(syn_y == 1)[0]

    real_idx = strat_samp(real_neg_i, real_pos_i, n_neg, n_pos)
    syn_idx  = strat_samp(syn_neg_i,  syn_pos_i,  n_neg, n_pos)

    real_s = real_X.iloc[real_idx]
    syn_s  = syn_X.iloc[syn_idx]

    X = pd.concat([real_s, syn_s], axis=0)
    s = np.r_[np.zeros(len(real_s), dtype=int), np.ones(len(syn_s), dtype=int)]

    rf = RandomForestClassifier(n_estimators=800, random_state=seed, n_jobs=-1)
    rf.fit(X, s)

    imp = pd.Series(rf.feature_importances_, index=X.columns).sort_values(ascending=False)
    return imp

imp = fit_rf_get_importance(real_X, real_y, syn_X, syn_y, n_neg, n_pos)
imp.head(20)


latex_str = imp.to_latex(
    index=False,
    float_format="%.3f",
    escape=True
)

with open("sifnificant.tex", "w", encoding="utf-8") as f:
    f.write(latex_str)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from matplotlib.patches import Ellipse
from matplotlib.lines import Line2D

# --------------------------------------------------
# Assume:
# oX     = real data matrix
# oy     = real labels (0/1)
# syn_X  = synthetic data matrix
# syn_y  = synthetic labels (0/1)
# --------------------------------------------------

# ---------- PCA ----------
pca = PCA(n_components=2, random_state=0)

Z_real = pca.fit_transform(oX)
Z_syn  = pca.transform(syn_X)

# ---------- ellipse helper ----------
def plot_ellipse(ax, X, color):
    cov = np.cov(X.T)
    mean = X.mean(axis=0)

    eigvals, eigvecs = np.linalg.eigh(cov)
    order = eigvals.argsort()[::-1]

    eigvals = eigvals[order]
    eigvecs = eigvecs[:, order]

    angle = np.degrees(np.arctan2(*eigvecs[:, 0][::-1]))
    width, height = 2 * 2 * np.sqrt(eigvals)

    ell = Ellipse(
        xy=mean,
        width=width,
        height=height,
        angle=angle,
        edgecolor=color,
        facecolor="none",
        lw=2
    )
    ax.add_patch(ell)

# ---------- colors by class ----------
class_colors = {
    0: "tab:blue",
    1: "tab:red"
}

# ---------- plot ----------
fig, ax = plt.subplots(figsize=(7, 6))

# Real points: circles
for cls in np.unique(oy):
    idx = (oy == cls)
    ax.scatter(
        Z_real[idx, 0],
        Z_real[idx, 1],
        c=class_colors[cls],
        marker="o",
        alpha=0.6,
        s=50
    )

# Synthetic points: triangles
for cls in np.unique(syn_y):
    idx = (syn_y == cls)
    ax.scatter(
        Z_syn[idx, 0],
        Z_syn[idx, 1],
        c=class_colors[cls],
        marker="^",
        alpha=0.6,
        s=50
    )

# Optional: ellipses by source only
plot_ellipse(ax, Z_real, "black")
plot_ellipse(ax, Z_syn, "gray")

ax.set_title("PCA projection by class and source")
ax.set_xlabel("PC1")
ax.set_ylabel("PC2")

# ---------- custom legend ----------
legend_elements = [
    Line2D([0], [0], marker='o', color='w', label='Real',
           markerfacecolor='gray', markeredgecolor='gray', markersize=8),
    Line2D([0], [0], marker='^', color='w', label='Synthetic',
           markerfacecolor='gray', markeredgecolor='gray', markersize=8),
    Line2D([0], [0], marker='o', color='w', label='Class 0',
           markerfacecolor='tab:blue', markeredgecolor='tab:blue', markersize=8),
    Line2D([0], [0], marker='o', color='w', label='Class 1',
           markerfacecolor='tab:red', markeredgecolor='tab:red', markersize=8),
]

ax.legend(handles=legend_elements, frameon=True)

plt.tight_layout()
plt.savefig("pca_class_source.png", dpi=300, bbox_inches="tight")
plt.show()
```

# Two-sample T-test

We want to test for each feature $j$:
$$
H_{0}: \mu_{\text{real}_{j}} = \mu_{\text{syn}_{j}}
$$
vs.
$$
H_{1}: \mu_{\text{real}_{j}} \ne \mu_{\text{syn}_{j}}
$$

```python
from scipy.stats import ttest_ind, f_oneway
import pandas as pd
import numpy as np
results = []
# X_real
X_syn = syn_X.to_numpy()
col_names = oX.columns
for j in range(X_real.shape[1]):
    
    real_vals = X_real[:, j]
    syn_vals  = X_syn[:, j]

    # Welch t-test (better when variances differ)
    t_stat, t_p = ttest_ind(real_vals, syn_vals, equal_var=False)

    # ANOVA
    f_stat, f_p = f_oneway(real_vals, syn_vals)

    results.append({
        "feature_index": j,
        "feature_name": col_names[j],
        "real_mean": real_vals.mean(),
        "syn_mean": syn_vals.mean(),
        "real_std": real_vals.std(),
        "syn_std": syn_vals.std(),
        "t_stat": t_stat,
        "t_p": t_p,
        "f_stat": f_stat,
        "f_p": f_p
    })

df_tests = pd.DataFrame(results)
```

```python
df_tests = df_tests.sort_values("t_p")

display(df_tests.head(10))
```

# FDR correction
63 hypothesis test sharing the same power ($\alpha = 0.05$). Addressing the multiple testing problem.

### Benjamini–Hochberg
1. Sort the p-values
2. Compare each to:
$$
\frac{i}{m}\alpha
$$
- $i$ = rank
- $m$ = number of tests

```python
from statsmodels.stats.multitest import multipletests

reject, p_adj, _, _ = multipletests(df_tests["t_p"], method="fdr_bh")

df_tests["p_adj"] = p_adj
df_tests["reject_H0"] = reject

display((df_tests[df_tests["p_adj"] < 0.05]).head(5))
```

```python
import matplotlib.pyplot as plt

diff = df_tests["real_mean"] - df_tests["syn_mean"]

plt.figure(figsize=(6,4))
plt.scatter(diff, -np.log10(df_tests["p_adj"]))
plt.xlabel("Mean difference (real - synthetic)")
plt.ylabel("-log10 adjusted p-value")
plt.axhline(-np.log10(0.05), linestyle="--", color="red")
plt.title("Feature-level differences: real vs synthetic")
for i in df_tests.nsmallest(5, "p_adj").index:
    plt.text(diff[i], -np.log10(df_tests["p_adj"][i]), df_tests["feature_name"][i])
plt.show()
```

We used mean difference because the t-test specifically tests differences in means.
- V10_blood_IgGRBD: the synthetic model is strongly underestimating this marker.
- V9_IL2: Mean difference smaller, but variance may be small, so the test still finds evidence of a difference.
